```python
import matplotlib.pyplot as plt
```

```python
import astropy
```

```python
import lightkurve as lk
```

```python
lk.__version__
```

```python
plt.rcParams['figure.figsize'] = (12, 4)
plt.rcParams['font.size'] = 10

print("Environment ready!")
```

```python
target="Kepler-10"
```

```python
search_result=lk.search_lightcurve(
    target,
    author="Kepler",
    cadence="long"
)
```

```python
print(f"Found {len(search_result)} light curve files for {target}")
print("Available quarters")
print(search_result)      
```

```python
lc=search_result[0].download()
```

```python
print(lc)
```

```python
print(f"Target ID: {lc.targetid}")
print(f"Number of data points: {len(lc.time)}")
print(f"Time range: {lc.time.min()} to {lc.time.max()}")
print(f"Flux range: {lc.flux.min()} to {lc.flux.max()}")
```

```python
plt.figure(figsize=(12,4))
plt.plot(
    lc.time.value,
    lc.flux.value,
    'k.',
    markersize=2,
    
)
plt.xlabel("Time(days)",fontsize=12)
plt.ylabel("Flux(e-/s)",fontsize=12)
plt.title("Kepler-10-Quarter0 Light curve",fontsize=14,fontweight="bold")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
start_time=lc.time.value[0]
end_time=start_time+5
```

```python
mask=(lc.time.value>=start_time)&(lc.time.value<=end_time)
plt.figure(figsize=(14,6))
plt.plot(
    lc.time.value[mask],
    lc.flux.value[mask],
    "k.",
    markersize=2
)
plt.xlabel("Time(days)")
plt.ylabel("Flux(e-/s)")
plt.title("Kepler-10- Zoomed in Light curve(first 5 days)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
